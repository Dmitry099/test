```python
1) Напишите программу, которая печатает числа от 1 до 100: для кратных трем вместо числа 
    печатать “Fizz” , для кратных пяти - печатать “Buzz”, для кратных и трем, и пяти - печатать 
    “FizzBuzz”
```

```python
for x in range (1, 100):
    if x % 3 == 0 and x % 5 == 0:
        x = 'FizzBuzz'
    elif x % 3 == 0:
        x = 'Fizz'
    elif x % 5 == 0:
        x = 'Buzz'
    print(x)
```
